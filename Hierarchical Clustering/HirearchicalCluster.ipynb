# Hierarchical Clustering (Aglomerative Clustering)

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Read the dataset

```python
import pandas as pd
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
df.info()
```

```python
df.columns = ['CustomerID','Gender','Age','Income','Spending']
```

```python
df
```

```python
df.isna().sum()
```

### Seperate X data

```python
X = df[['Age','Income','Spending']]
X
```

### Visulize the data

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

x = X['Age']
y = X['Income']
z = X['Spending']

ax.scatter(x,y,z)

plt.show()
```

### Preprocess the data with sklearn

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre[0:5]
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre,columns=cols)
```

```python
X_pre.head()
```

### Silhoutee method to calculate optimum number of clusters

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
min_clusters = 2
max_clusters = 15
cluster_numbers = []
silhoutee = []

for i in range(min_clusters,max_clusters+1):
    # Call AGC model
    agc = AgglomerativeClustering(n_clusters=i)

    # Fit predict the labels
    labels = agc.fit_predict(X_pre)

    # Calculate silhoutte score
    sil = silhouette_score(X_pre,labels)
    silhoutee.append(sil)

    cluster_numbers.append(i)
```

### Visualising the silhoutte score

```python
plt.plot(cluster_numbers,silhoutee,marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhoutee Score')
plt.title('Silhoutee Analysis')
plt.show()
```

### Create final algo with 6 clusters because highest siloutee score is for 6 clusters

```python
model = AgglomerativeClustering(n_clusters=6)
model.fit(X_pre)
```

```python
labels = model.labels_
labels
```

```python
X['label']=labels
```

```python
X
```

### View the clusters in Dataframe format

```python
for i in range(6):
    print(f'Label No : {i}')
    display(X[X['label']==i])
    print('\n================================================\n')    
```

### Visulize the clusters

```python
def plotcolor(ar):
    col = []
    for i in ar:
        if i==0:
            col.append('red')
        elif i==1:
            col.append('green')
        elif i==2:
            col.append('blue')
        elif i==3:
            col.append('yellow')
        elif i==4:
            col.append('violet')
        else:
            col.append('orange')
    return col
```

```python
cl = plotcolor(labels)
print(cl)
```

### Visualize the clusters

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = X['Age']
y = X['Income']
z = X['Spending']

ax.scatter(x,y,z,c=cl)

plt.show()
```

### Dendogram 

```python
from scipy.spatial import distance_matrix
DM = distance_matrix(X_pre,X_pre)
from scipy.cluster.hierarchy import linkage
linkages = linkage(DM)
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(16,8))
Q = dendrogram(linkages)
plt.show()
```
