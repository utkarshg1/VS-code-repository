### Renaming column names for simplicity

```python
df.columns = ['CustomerId','Gender','Age','AnnualIncome','SpendingScore']
df.head()
```

### Clustering customers based on Age, AnnualIncome and SpendingScore

```python
X = df[['Age','AnnualIncome','SpendingScore']]
X.head()
```

### Visualize the data in 3d

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

x = X['Age']
y = X['AnnualIncome']
z = X['SpendingScore']

ax.scatter(x,y,z)
plt.show()
```

### Scaling the data

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre
```

```python
colms = num_pipe.get_feature_names_out()
colms
```

### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
agc = AgglomerativeClustering(n_clusters=6)
agc.fit(X_pre)
```

```python
lb = agc.labels_
lb
```

### Dendogram visualization

```python
from scipy.spatial import distance_matrix
DM = distance_matrix(X,X)
from scipy.cluster.hierarchy import linkage
linkages = linkage(DM)
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title('Dendogram')
Q = dendrogram(linkages)
```

### Defining unique color for each cluster

```python
def plotcolor(ar):
    cols = []
    for i in ar:
        if i==0:
            cols.append('red')
        elif i==1:
            cols.append('green')
        elif i==2:
            cols.append('blue')
        elif i==3:
            cols.append('yellow')
        elif i==4:
            cols.append('violet')
        else:
            cols.append('orange')
    return cols
```

```python
cols = plotcolor(lb)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = X['Age']
y = X['AnnualIncome']
z = X['SpendingScore']

ax.scatter(x,y,z,c=cols)
plt.show()
```

```python
min_clusters=2
max_clusters=15
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
# Initialize lists to store silhouette scores and cluster numbers
silhouette_scores = []
cluster_numbers = []

for i in range(min_clusters,max_clusters+1):
    # Initialize the hierarchical clustering algorithm
    clustering = AgglomerativeClustering(n_clusters=i)
    
    # Fit the algorithm to your data
    labels = clustering.fit_predict(X_pre)
    
    # Calculate the silhouette score
    silhouette_avg = silhouette_score(X_pre, labels)
    
    # Append the silhouette score and cluster number to the lists
    silhouette_scores.append(silhouette_avg)
    cluster_numbers.append(i)
```

```python
# Plot the silhouette scores against the number of clusters
plt.plot(cluster_numbers, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis')
plt.show()
```
